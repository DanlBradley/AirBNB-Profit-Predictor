```python
bnb_listings['price'].fillna(0, inplace=True)

#Function to convert string money to float money
def money_str_to_float(x):
    if type(x) == str:
        x = x.replace('$','')
        x = x.replace(',','')
        return float(x)
    elif type(x) == int:
        return float(x)
    
bnb_listings['price'] = bnb_listings['price'].apply(lambda x: money_str_to_float(x))    
bnb_listings['price'].quantile(0.95)
```

```python
bnb_listings['price']
```

```python
quan = bnb_listings['price'].quantile(0.95)
bnb_listings.loc[bnb_listings['price'] >= quan].index
```

```python
def outliers(column, n = .95):
    quan = column.quantile(n)
    outliers = column.loc[column >= quan].index
    return outliers
```

```python
outliers(bnb_listings['price'])
```

```python
df_columns = ['bedrooms', 'host_listings_count', 'price', 'minimum_nights']
drop = []
for each_column in df_columns:
    remove = list(outliers(bnb_listings[each_column],0.99))
    for item in remove:
        drop.append(item)
len(set(drop))
```

```python
import numpy as np

bnb_listings.drop(labels = (np.array(drop)),axis = 0)
```
